# 第一节 数据载入

数据载入有相对路径和绝对路径，最好用绝对路径，可以看到文件的存储位置，
读取csv文件可以用read_csv和read_table两种，区别在于分割方式不一样，读csv文件用read_csv更方便
数据量太大的时候，用逐块读取的方式更高效

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df=pd.read_csv('train.csv')
df.rename(columns={'PassengerId':'乘客ID','Survived':'是否幸存','Pclass':'乘客等级','Name':'乘客姓名','Sex':'性别','Age':'年龄','SibSp':'兄弟姐妹数量','Parch':'父母子女数量','Ticket':'船票信息','Fare':'票价','Cabin':'客舱','Embarked':'登船港口'},inplace=True)
df=df.set_index('乘客ID')
df.head()
```

# 第二节 pandas基础

pandas处理数据的时候要注意索引的问题，筛选数据之后索引会变，当提取部分数据的时候会出现问题，因此要重新设置索引

```python
midage=df[(df['年龄']>10)&(df['年龄']<50)]
midage=midage.reset_index(drop=True)
midage.head()
```

提取数据可以用loc和iloc,loc是通过标签提取，iloc是通过位置提取

```python
midage.loc[[100],['乘客等级','性别']]
```

```python
midage.iloc[[100],[1,3]]
```

# 第三节 探索性数据分析

通过排序进行分析

```python
df.sort_values(by=['票价','年龄'],ascending=False).head(10)#ascending=False 降序，True升序
```

DataFrame 进行算术运算

```python
max(df['兄弟姐妹数量'] + df['父母子女数量'])
```

```python
df['sum']=df['兄弟姐妹数量']+df['父母子女数量']
df.sort_values(by=['sum'],ascending=False).head(1)
```

```python
sum(df[df['性别']=='female']['是否幸存']),sum(df[df['性别']=='male']['是否幸存']) #可以看出女性存活数量多于男性
```
